# Sales Transaction
- The Sales Jan 2009 file contains some “sanitized” sales transactions during the month of January. Below is a sample of a report built in just a couple of minutes using the Blank Canvas app. These 998 transactions are easily summarized and filtered by transaction date, payment type, country, city, and geography.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
data = pd.read_csv("SalesJan2009.csv")
data.info()
```

```python
data
```

```python
data.describe()
```

```python
data['Price'].value_counts()
```

```python
data['Payment_Type'].unique()
```

```python
df_price = pd.DataFrame(data, columns=["Product","Price"])
```

```python
df_price.groupby(["Product"])[["Price"]].describe()
```

```python
df_product = df_price.pivot_table(index='Product')
df_product
```

```python
df_product.plot(kind='bar', figsize=(10,8))
plt.title('Price per product')
plt.show()
```
